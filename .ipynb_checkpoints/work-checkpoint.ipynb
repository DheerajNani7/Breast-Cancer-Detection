# Data Preprocessing

```python
d['diagnosis'] = d['diagnosis'].map({'M':1,'B':0})
```

```python
d
```

```python
y= d["diagnosis"]
x=d.drop("diagnosis",1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=9)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_train[0]
```

```python
from sklearn.decomposition import PCA

pca = PCA()
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
e_v = pca.explained_variance_ratio_
```

```python
e_v
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=13)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

# Building various models

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)
classifier.fit(x_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('\n')
print( accuracy_score(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (2, 2))
sns.heatmap(cm, annot = True, cbar = False)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
# to find which value shows the lowest mean error
error_rate = []

for i in range(2,42):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(2,42), error_rate, color='red', linestyle="--",
         marker='o', markersize=10, markerfacecolor='b')
plt.title('Error_Rate vs K-value')
plt.show()
```

#  Ensembling the models (Voting and Stacking) 

```python
# importing machine learning models for prediction
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# importing voting classifier
from sklearn.ensemble import VotingClassifier

# Making the final model using voting classifier
final_model = VotingClassifier(estimators=[('rf', classifier), ('knn', knn), ('dt', dt), ('nb', gnb), ('svm', clf)], voting='hard')

# training all the model on the train dataset
final_model.fit(x_train, y_train)


# predicting the output on the test dataset
y_pred = final_model.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('\n')
print( accuracy_score(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression() # defining meta-classifier
clf_stack = StackingClassifier(estimators = [('rf', classifier), ('knn', knn), ('dt', dt), ('nb', gnb), ('svm', clf)],
                                       final_estimator = lr)
```

```python
from sklearn import model_selection
print('Doing 3-fold cross validation:\n')

for iterclf, iterlabel in zip([classifier,knn,dt,gnb,clf,clf_stack],
                             ['Random Forest',
                              'k-Nearest Neighbours',
                              'Decision Tree',
                              'Naive Bayes',
                              'Support Vector Machine',
                              'Stacking Classifier'
                             ]):
    scores = model_selection.cross_val_score(iterclf, x_train, y_train, cv = 3, scoring='accuracy')
    print("Accuracy: %0.3f (+/- %0.3f) [%s]" % (scores.mean(), scores.std(), iterlabel))
```

# ANN model

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Dropout


#CREATING THE ANN AS SEQUENCE OF LAYERS
ann = tf.keras.models.Sequential()

#ADDING FIRST HIDDEN LAYER WITH 13 NEURONS, THE INPUT LAYER WILL BE ADDED AUTOMATICALLY,
ann.add(tf.keras.layers.Dense(units = 13,activation = 'relu'))
ann.add(Dropout(0.5))

#ADDING 2ND HIDDEN LAYER WITH 13 NEURONS
ann.add(tf.keras.layers.Dense(units = 13,activation = 'relu'))
ann.add(Dropout(0.5))

#ADDING OUTPUT LAYER WITH 1 NEURON , AS THIS IS A BINARY CLASSIFICATION
ann.add(tf.keras.layers.Dense(units = 1,activation = 'sigmoid'))

#COMPILING THE ANN USING STOCHASTIC GRADIENT DESCENT (optimizer = 'adam')
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early = EarlyStopping(monitor = 'accuracy', mode = 'max',patience = 25)
```

```python
history=ann.fit(x_train, y_train,batch_size = 32, epochs = 100,callbacks = [early])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred = ann.predict(x_test)
y_pred  = y_pred > 0.8
cm = confusion_matrix(y_pred,y_test)
ac = accuracy_score(y_pred,y_test)
```

```python
import seaborn as sns
plt.figure(figsize = (15,10))
sns.heatmap(cm,annot = True,annot_kws={'size': 25})
plt.title('CONFUSION MATRIX')
print('Accuracy - {}%'.format(ac*100))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('\n')
print( accuracy_score(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```
