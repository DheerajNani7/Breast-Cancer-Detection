Data Visualization

```python
plt.figure(figsize = (8,7))
sns.countplot(x="Type", data=d, palette='magma')
```

```python
d['Type'].count()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

g = sns.PairGrid(d.iloc[:, 0:11], hue="Type", palette="Set2")
g = g.map_diag(plt.hist, edgecolor="w")
g = g.map_offdiag(plt.scatter, edgecolor="w", s=40)
plt.show()
```

```python
df_corr = d.iloc[:, 1:11].corr()
plt.figure(figsize=(8,8))
sns.heatmap(df_corr, cmap="Blues", annot=True)
plt.show()
```

```python
df_corr = d.iloc[:, 11:21].corr()
plt.figure(figsize=(8,8))
sns.heatmap(df_corr, cmap="Blues", annot=True)
plt.show()
```

```python
df_corr = d.iloc[:, 21:].corr()
plt.figure(figsize=(8,8))
sns.heatmap(df_corr, cmap="Blues", annot=True)
plt.show()
```
